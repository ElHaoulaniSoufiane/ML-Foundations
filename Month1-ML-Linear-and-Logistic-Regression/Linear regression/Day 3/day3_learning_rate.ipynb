# Day 3 — Learning Rate & Convergence

yesterday gradient descent found w automatically  
today: what happens if the step size (alpha) is too big or too small  

this is important because a bad alpha can break everything even if the rest of the code is correct

```python
import matplotlib.pyplot as plt
```

```python
x_train = [1, 2, 3, 4, 5]
y_train = [2, 4, 6, 8, 10]  # y = 2x, perfect w = 2

# we'll test these learning rates to see what each one does
alphas = [0.001, 0.01, 0.1, 0.5, 1.0]

epochs = 100
```

```python
def compute_cost(x, y, w, b=0):
    m = len(x)
    total = 0
    for i in range(m):
        pred = w * x[i] + b
        total += (pred - y[i]) ** 2
    return (1 / (2 * m)) * total


def compute_gradients(x, y, w, b=0):
    m = len(x)
    dw = 0
    for i in range(m):
        pred = w * x[i] + b
        dw += (pred - y[i]) * x[i]
    return dw / m


def gradient_descent(x, y, w, alpha, epochs):
    cost_history = []
    for _ in range(epochs):
        dw = compute_gradients(x, y, w)
        w = w - alpha * dw
        cost = compute_cost(x, y, w)
        cost_history.append(cost)
        # stop early if cost explodes — means alpha is too big
        if cost > 1e10:
            break
    return w, cost_history
```

```python
# run gradient descent for each alpha and store results
results = {}

for alpha in alphas:
    w_final, cost_history = gradient_descent(x_train, y_train, w=0.0, alpha=alpha, epochs=epochs)
    results[alpha] = {
        'w_final': w_final,
        'cost_history': cost_history,
        'final_cost': cost_history[-1]
    }
    print(f"alpha={alpha} → final w={w_final:.4f}, final cost={cost_history[-1]:.6f}")
```

```python
# plot cost curves for all alphas side by side
# this is the key visual — see how each alpha behaves differently
fig, axes = plt.subplots(1, len(alphas), figsize=(18, 4))

for ax, alpha in zip(axes, alphas):
    history = results[alpha]['cost_history']
    w_final = results[alpha]['w_final']

    ax.plot(history, color='royalblue', linewidth=2)
    ax.set_title(f'alpha = {alpha}\nfinal w = {w_final:.3f}', fontsize=11)
    ax.set_xlabel('epoch')
    ax.set_ylabel('cost')
    ax.grid(True, alpha=0.3)

plt.suptitle('Effect of Learning Rate on Convergence', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# zoom in on just the good ones to compare speed of convergence
good_alphas = [0.001, 0.01, 0.1]

plt.figure(figsize=(8, 5))
for alpha in good_alphas:
    plt.plot(results[alpha]['cost_history'], label=f'alpha={alpha}', linewidth=2)

plt.title('Convergence speed — smaller alpha = slower')
plt.xlabel('epoch')
plt.ylabel('cost')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# what divergence looks like — alpha too big
# the cost goes up instead of down
plt.figure(figsize=(7, 4))
bad_history = results[1.0]['cost_history']
plt.plot(bad_history, color='red', linewidth=2)
plt.title('alpha=1.0 — divergence (cost explodes)')
plt.xlabel('epoch')
plt.ylabel('cost')
plt.grid(True, alpha=0.3)
plt.show()
```

## what i learned today

- alpha too small → converges but takes forever
- alpha too big → diverges, cost goes up instead of down, useless
- alpha just right → converges fast and reaches low cost
- there's no perfect formula for alpha — you test and adjust
- in real ML, techniques like learning rate scheduling adjust alpha automatically during training
